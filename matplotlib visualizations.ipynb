```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize = (11, 6))

plt.title('name')
plt.plot(x, x**2)
plt.plot(x, -1 * (x **2))
```

```python
plt.figure(figsize = (15, 8))
plt.title('training')

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'V Line'])
plt.xlabel('X')
plt.ylabel('X^2')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'H Line'])
plt.xlabel('X')
plt.ylabel('-X^2')
```

### OOP Visualizations

```python
fig, axes= plt.subplots(figsize = (12,6))
```

```python
axes.plot(
    x, x**2, color = 'red', linewidth = 3,
    marker = 'o', markersize = 8, label = 'X^2')

axes.plot(x, -1 *(x**2), 'b--', label = '-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title('a plot')
axes.legend()

fig
```

```python
fig, axes = plt.subplots( figsize = (15, 6))

axes.plot(x, x, linestyle = 'solid')
axes.plot(x, x+1, linestyle = 'dashed')
axes.plot(x, x+2, linestyle = 'dashdot')
axes.plot(x, x+3, linestyle = 'dotted')

axes.set_title('another plot')
```

```python
fig, axes = plt.subplots( figsize = (15, 6))

axes.plot(x, x, '-og', label = 'solid green')
axes.plot(x, x+1, '--c', label = 'dashed cyan')
axes.plot(x, x+2, '-.b', label = 'dashdo tblue')
axes.plot(x, x+3, ':r', label = 'dotted red')

axes.legend()
axes.set_title('another plot')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * (20 *np.random.rand(N)) **2
```

```python
plt.figure( figsize = (14,6))

plt.scatter(x, y, s = area, c = colors, alpha = 0.5, cmap = 'Spectral')
plt.colorbar()
```

```python
fig = plt.figure(figsize = (14, 6))

ax1 = fig.add_subplot(1,2,1)
plt.scatter(x, y, s = area, c = colors, alpha = 0.5, cmap = 'Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1,2,2)
plt.scatter(x, y, s = area, c = colors, alpha = 0.5, cmap = 'Pastel2')
plt.colorbar()

plt.show()
```

```python
vals = np.random.randn(1000)
```

```python
plt.subplots( figsize = (14, 6))

plt.hist(vals, bins = 100, alpha = 0.8,
         histtype = 'bar', color = 'steelblue',
         edgecolor = 'green')
plt.xlim(xmin = -5, xmax = 5)
plt.show()
```

```python
fig.savefig('hist.png')
```

```python
from scipy import stats
density = stats.kde.gaussian_kde(vals)
density
```

```python
plt.subplots(figsize= (14,6))

vals2 = np.linspace(min(vals)-10, max(vals)+10, 100)

plt.plot(vals2, density(vals2), color = 'red')
plt.fill_between(vals2, 0, density(vals2), alpha = 0.5, color = 'red')
plt.xlim(xmin = -5, xmax = 5)

plt.show()
```

```python
plt.subplots(figsize= (14,6))

vals2 = np.linspace(min(vals)-10, max(vals)+10, 100)

plt.hist(vals, bins = 100, alpha = 0.8,
         histtype = 'bar', color = 'steelblue',
         edgecolor = 'green')

plt.plot(vals2, density(vals2), color = 'red', linewidth = 3.0)

plt.xlim(xmin = -5, xmax = 5)

plt.show()
```

```python
Y = np.random.rand(1,5)[0]
Y2 = np.random.rand(1,5)[0]
```

```python
plt.figure(figsize = (14,6))

barwidth = 0.5
plt.bar(np.arange(len(Y)), Y, width = barwidth, color = 'steelblue')

plt.show()
```

```python
plt.figure(figsize=(14,6))

barw = 0.5
plt.bar(np.arange(len(Y)), Y, width= barw, color = 'steelblue', label = 'label Y')
plt.bar(np.arange(len(Y2)), Y2, width = barw, color = 'green', bottom = Y, label = 'label Y2')

plt.legend()
plt.show()
```
