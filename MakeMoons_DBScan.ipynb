```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.decomposition import PCA
```

```python
df = pd.read_pickle('dbscan2000.pkl')
plt.scatter(df[:, 0], df[:, 1], s = 10)
data = pd.DataFrame(dict(X = df[:, 0], Y = df[:, 1]))
data = data.values
```

```python
df = pd.read_csv('Clustering_gmm.csv')
```

```python
epsilon = 1.0
min_samples = 1
dbscan = DBSCAN(eps = epsilon, min_samples = min_samples)
dbscan.fit(data)
```

```python
cluster_labels = dbscan.labels_
```

```python
pca = PCA(n_components = 2)
data_pca = pca.fit_transform(data)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c = cluster_labels, cmap = 'rainbow', marker = 'o', s = 50)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('DBSCAN Clustering with PCA Visualization')
plt.show()
```
